```python
from google.colab import drive
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical



# Set dataset path (Update this with your actual path)
dataset_path = '/content/drive/MyDrive/criminal_dataset'

# Image size and number of classes (adjust as needed)
IMG_SIZE = 128
NUM_CLASSES = 10  # Change based on the number of unique criminals

# Load images and labels
data = []
labels = []
for category in os.listdir(dataset_path):
    category_path = os.path.join(dataset_path, category)
    if os.path.isdir(category_path):
        for img_name in os.listdir(category_path):
            img_path = os.path.join(category_path, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            data.append(img)
            labels.append(int(category))  # Assuming folder names represent class labels

data = np.array(data).reshape(-1, IMG_SIZE, IMG_SIZE, 1) / 255.0
labels = np.array(labels)

# Convert labels to one-hot encoding
labels = to_categorical(labels, NUM_CLASSES)

# Split dataset
train_X, test_X, train_Y, test_Y = train_test_split(data, labels, test_size=0.2, random_state=42)
```
